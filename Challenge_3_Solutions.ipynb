```python
#To Do !!
#Create a to-do list. It should at least be able to do the following things:
#Add an item to the To-Do list
#Delete an item from the To-Do list
#Print the To-Do list
#All the To-Do's should be written to a file so that you can load it later.

import re

def addTask(comm, tsk_lst):              #defining functions
    tsk = comm.split(sep='\'')[1]      #dividing text by ' => addTask('Do smth') -->split-->  ['addTask(', 'Do smth', ')']
                                        #only writing 1st element to  tsk
    tsk_lst.append(tsk)                 #adding that element to our list


def delTask(comm, tsk_lst):
    num_tsk = comm[8:len(comm)-1]       #delTask(someNumber)
    num_tsk = int(num_tsk)
    if num_tsk > len(tsk_lst):          #checking if the number exists
        print("number out of list")
    else:
        tsk_lst.pop(num_tsk-1)              #deleting number from list


def viewTask(tsk_lst):
    if len(tsk_lst) == 0:
        print("task list is empty")
    else:
#       for i in range(len(tsk_lst)):
#          print('{}. {}'.format(i+1, tsk_lst[i]))     #alternative printing the tasks with item  number
        for num, task in enumerate(tsk_lst, start=1):
            print('{}. {}'.format(num, task))

commands = [r'addTask\(\'[A-Za-z0-9\s]+\'\)', r'delTask\([1-9]+[0-9]*\)', r'viewTask\(\)', r'exit\(\)']    #reg exp-inputing the task

tasks = []                              #preparation for working with task

try:                                    #trying to open a file
    fh = open('todolist.dat', 'r')      #fh = file handler. We open file to read
    for st in fh:
        tasks.append(st.replace('\n', ''))    #adding earlier created tasks from file to list and deleting symbol 
        #of transfer to another line
except FileNotFoundError:
    fh = open('todolist.dat', 'w')      #except if the file does not exist =>making a new file and write in it
finally:
    fh.close()                          #closing file. it is guaranteed that it exists now.

print ("input the instruction:\n"       #giving instructions to the user 
       "Add an item to the To-Do list\n"
       "Delete an item from the To-Do list\n"
       "Print the To-Do list")


while True:                             #infinite cycle
    inst = input()
    #print(re.match(patt, inst))
    if re.match(commands[0], inst) is not None:     #special construction for regular expressions:check if input matches
        addTask(inst, tasks)
    elif re.match(commands[1], inst) is not None:
        delTask(inst, tasks)
    elif re.match(commands[2], inst) is not None:
        viewTask(tasks)
    elif re.match(commands[3], inst) is not None:
        break
    else:
        print('Wrong command')


with open('todolist.dat', 'w') as fh:   #context manager, open file in writing mode
    for task in tasks:
        fh.write(task+'\n')             #writing task to file and adding transfer to the new line
```

```python
#Find your "weight" on other Planets!!

from math import pi

G = 6.67e-11


def force_calc(m, planet_info):
    R = planet_info[1]
    V = 4/3*pi*R**3
    D = planet_info[2]
    M = V*D
    F = G*M*m/R**2
    return F

m = input("input mass of the object ")
m = int(m)

n = input("input the total number of planets ")
n = int(n)

planets = []
for i in range(n):
    planet_info = input()
    planet_info = planet_info.split(sep=", ")
    planet_info[1] = int(planet_info[1])
    planet_info[2] = int(planet_info[2])
    planets.append(planet_info)

for planet in planets:
    weight = force_calc(m, planet)
    print('{}: {:.3f}'.format(planet[0], weight))  #:.3f - for the formating 3 digits after comma
```
