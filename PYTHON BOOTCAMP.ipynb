```python
#lets make x and y, and make some noise for y
x = np.arange(100)
y = np.random.randn(100) + np.sqrt(x)
```

```python
#by definition it prints in the last plot
fig = pl.figure(figsize=(13,4))
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)
pl.plot(x,y)
```

```python
#lets now plot using ax functions and ay functions
fig = pl.figure(figsize=(13,4))
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)
ax1.plot(x,y, label = 'my sq root')
ax1.legend(fontsize=15)
ax2.plot(x,y, '.-', label = 'my sq root')
```
