```python
#1. create the vector v=(3,7) and plot it.
# importing future for compatibility between python 2 and python 3
from __future__ import print_function

# importing needed packages
import numpy as np
import pylab as pl

# The seaborn package makes plots prittier
import seaborn as sns
sns.set_style("white")
# The dollowing command (it is called a magic command) 
# is needed to display plots in the notebook.
%pylab inline
#Populating the interactive namespace from numpy and matplotlib

v = np.array([[3], [7]])

ax = pl.axes()
ax.arrow(0, 0, v[0][0], v[1][0], head_width = 0.3, head_length = 0.7)
pl.xlim(-14, 14)
pl.ylim(-14, 14)
pl.grid()


pl.plot()
```

```python
#2. Modify vector v so as to get a vector twice as long, in the same direction of v=(3,7). 
#Call the new vector vLong and plot it.
#For this and all following plots also plot the orivinal vector v with a dashed, grey line.

vLong = v*2
ax = pl.axes()
ax.arrow(0, 0, vLong[0][0], vLong[1][0], head_width = 0.3, head_length = 0.7, color = 'blue')
ax.arrow(0, 0, v[0][0], v[1][0], head_width = 0.3, head_length = 0.7, color = 'grey',  ls = 'dashed', lw = 2)
pl.xlim(-14, 14)
pl.ylim(-14, 20)
pl.grid()

pl.plot()
```

```python
#3. Create a function, that takes as argument an angle (in degrees) and returns a roatation matrix (also a numpy ndarray).
#Name the function rotMatrix.

#4. apply the rotation matrix to the vector v=(3,7) and rotate it by 30 degrees counterclockwise, call this vector v30, 
#and then by 90 degrees clockwise call this vector v90. plot them.

def rotMatrix(angle):
    angle = angle/180*np.pi
    matr = np.array([[np.cos(angle), -np.sin(angle)], [np.sin(angle), np.cos(angle)]])
    return matr
matr30 = rotMatrix(30)
v30 = matr30.dot(v)
matr90 = rotMatrix(-90)
v90 = matr90.dot(v)
ax = pl.axes()
ax.arrow(0, 0, v30[0][0], v30[1][0], head_width = 0.3, head_length = 0.7, color = 'blue')
ax.arrow(0, 0, v90[0][0], v90[1][0], head_width = 0.3, head_length = 0.7, color = 'red')
ax.arrow(0, 0, v[0][0], v[1][0], head_width = 0.5, head_length = 1, color = 'grey',  ls = 'dashed', lw = 2)
pl.axis('equal')
pl.xlim(-14, 15)
pl.ylim(-14, 15)
pl.grid()


pl.plot()
```

```python
#5. Sum the vectors vLong and v90. Plot the original vectors with dashed grey lines and the resulting vector.
vSum = v90 + vLong
ax = pl.axes()
ax.arrow(0, 0, vSum[0][0], vSum[1][0], head_width = 0.3, head_length = 0.7, color = 'blue')
ax.arrow(0, 0, v90[0][0], v90[1][0], head_width = 0.3, head_length = 0.7, color = 'grey', ls = 'dashed')
ax.arrow(0, 0, vLong[0][0], vLong[1][0], head_width = 0.3, head_length = 0.7, color = 'grey',  ls = 'dashed')
#pl.axis('equal')
pl.xlim(-14, 15)
pl.ylim(-19, 20)
pl.grid()


pl.plot()
```
