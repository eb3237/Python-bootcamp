```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import pandas as pd
import datetime
from pprint import pprint

from itertools import cycle
def mplcolors():
    fav_colors = [
        '#1abc9c', '#2ecc71', '#3498db', '#9b59b6', '#34495e',
        '#16a085', '#27ae60', '#2980b9', '#8e44ad', '#2c3e50',
        '#f1c40f', '#e67e22', '#e74c3c', '#ecf0f1', '#95a5a6',
        '#f39c12', '#d35400', '#c0392b', '#bdc3c7', '#7f8c8d'
    ]
    return cycle(fav_colors)

cycled_colors = mplcolors()

def get_hour(st):
    return datetime.datetime.strptime(st, '%Y-%m-%d %H:%M:%S').hour


plt.style.use('seaborn-darkgrid')
# df = pd.read_csv('data.csv')
df = pd.read_csv('201707-citibike-tripdata.csv')


mean_time = df['tripduration'].mean()
print('mean_time =', mean_time, 'seconds' )


df['hour'] = df['starttime'].apply(get_hour)

stat = [[], [], []]

for gend in range(3):
    df_ind = df[df['gender'] == gend]
    for h in range(24):
        stat[gend].append(len(df_ind[df_ind['hour'] == h]))

#pprint(stat)


male = len(df[df['gender'] == 1])
female = len(df[df['gender'] == 2])
print('male num =', male)
print('female num =', female)

subscriber = len(df[df['usertype'] == 'Subscriber'])
customer = len(df[df['usertype'] == 'Customer'])
print('subscriber num =', subscriber)
print('customer num =', customer)


df_gend = pd.DataFrame(stat).transpose()
df_gend['sum'] = df_gend[0]+df_gend[1]+df_gend[2]


fig = plt.figure(figsize=(15,8))
gs = gridspec.GridSpec(2,2)
ax1 = fig.add_subplot(gs[0, :])
ax2 = fig.add_subplot(gs[1, 0])
ax3 = fig.add_subplot(gs[1, 1])


ax1.bar(np.arange(24), df_gend[1], color = next(cycled_colors), label = 'Men', edgecolor = 'black', lw = 0.5, align = 'edge')
ax1.bar(np.arange(24), df_gend[2], color = next(cycled_colors), label = 'Women', bottom = df_gend[1], edgecolor = 'black', lw = 0.5, align = 'edge')
ax1.bar(np.arange(24), df_gend[0], color = next(cycled_colors), label = 'Unknown', bottom = df_gend[1]+df_gend[2], edgecolor = 'black', lw = 0.5, align = 'edge')
ax1.legend()

ax1.plot(np.arange(0.5, 24, 1), df_gend['sum'], color = next(cycled_colors))
ax1.set_xticks(np.arange(24))


hist_usage, hist_edges = np.histogram(df['hour'], bins=np.arange(0,25,4))
print(hist_usage)
print(hist_edges)

bars = ax1.bar(hist_edges[:-1], hist_usage//4, edgecolor = 'black', lw = 0.5, align = 'edge', alpha = 0.5, width = 4)
for thispatch in bars.patches:
    thispatch.set_facecolor(next(cycled_colors))

max_p = 1500000

ax2.bar([0, 1], [male, female], alpha = 0.5)
ax2.set_xticks([0, 1])
ax2.set_xticklabels(['Male', 'Female'])
ax2.set_ylim(max_p)

ax3.bar([0, 1], [subscriber, customer], alpha = 0.5)
ax3.set_xticks([0, 1])
ax3.set_xticklabels(['Subscriber', 'Customer'])
ax3.set_ylim(max_p)

plt.show()

```
